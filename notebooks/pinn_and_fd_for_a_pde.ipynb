## PINN & Finite Differences for 1D Wave Equation

Using a Physics-Informed Neural Network (PINN) as an ansatz and demonstrating a traditional numerical appraoch (finite-differences) to solving a partial differential equation (PDE).

The PDE is: \
    $ u_{tt}=c^2 u_{xx} $ with $u(0,t)=u(L,t)=0$ and initial conditions $ u(x,0)=\sin(\pi x) $ and $ u_t(x,0)=0 $.

```python
import pinn.pinn
```

```python
# Finite Difference Simulation
fd_solver = pinn.pinn.FiniteDifferenceWaveSolver(L=1.0, c=1.0, T=1.0, nx=101, nt=251)
fd_solver.run()
fd_solver.plot(step=25)
```

```python
# PINN Simulation
pinn_solver = pinn.pinn.WaveEquationPINNSolver(L=1.0, T=1.0, c=1.0)
pinn_solver.train(n_epochs=5000, lr=1e-3)
pinn_solver.plot(n_plot=101, time_stride=20)
```
