## Dynamic mode decomposition for nonlinear dynamics demonstration
Approximation of the Koopman operator.

```python
import koopman.dmd
import koopman.simulate
import koopman.visualizations
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
# Example TimeSeries
timeseries = koopman.simulate.TimeSeries()

steps = 100
timeseries.forward(timesteps=100)

timeseries.plot()
```

```python
timeseries = koopman.simulate.TimeSeries()

steps = 100
timeseries.forward(steps)

A, eigenvalues, eigenvectors = koopman.dmd.dmd(timeseries.state_history)

print("Koopman operator:\n", A)
print("Eigen values:\n", eigenvalues)
print("Eigen vectors:\n", eigenvectors)
```

```python
koopman.visualizations.plot_original_vs_dmd_reconstruction(timeseries.state_history, A, eigenvalues, eigenvectors)
```

```python
koopman.visualizations.plot_eigenvalues(eigenvalues)
```

```python
koopman.visualizations.plot_modes(eigenvectors)
```
